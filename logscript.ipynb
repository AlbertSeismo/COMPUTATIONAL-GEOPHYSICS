```python
import matplotlib
from matplotlib.pyplot import *
%matplotlib notebook
from scipy import *
from numpy import *
from pandas import *
```

```python
data = read_csv(r'C:\Users\akabanda\Desktop\DATA ANALYSIS-PYTHON\WELL LOGS\6507_11_8 Sw & Phi3.csv')
```

```python
data
```

```python
datafr = DataFrame(data)
```

```python
x = datafr['DEPTH']
y = datafr['SW']
z = datafr['PHIE']
```

```python
figure(figsize=(15,5))
plot(x,y,'r', linewidth =1.5)
grid(True)
xlabel('Depth/ft')
ylabel('Water Saturation/(v/v)')
title('Variation of Water Saturation with depth')
ylim([0,1.05])
show()
```

```python
figure(figsize=(15,5))
plot(x,z,'g',linewidth =1.5)
xlabel('Depth/ft')
ylabel('Porosity/(v/v)')
title('Variation of porosity with depth')
grid(True)

show()
```

```python
figure(figsize=(15,5))
plot(x,z,'g',x,y,'r',linewidth =1.5)
xlabel('Depth/ft')
ylabel('Porosity & Saturation')
title('Variation of porosity/water saturation with depth')
legend([['Porosity'],['Water Saturation']], loc ='best')
grid(True)

show()
```
