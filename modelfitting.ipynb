```python
#lets plot saturation and porosity vs depth
x = las['DEPTH']
y = las['SW']
z = las['PHIE']
```

```python
%matplotlib notebook

figure(figsize= (8,8))
subplot(1,3,1)
plot(y,x, LineWidth = 2)
gca().invert_yaxis()
xlim([0,1.1])
xlabel('Water Saturation/(V/V)')
ylabel('Depth/(V/V)')
grid()

subplot(1,3,3)
plot(z,x, LineWidth = 2)
gca().invert_yaxis()
#xlim([0,1.1])
grid()
xlabel('Porosity/(V/V)')
ylabel('Depth/(V/V)')
title('The figure below shows a plot of Water \nsaturation and porosity vs depth\n')
```
